# Variable transformation and Outlier Treatment

```python
final['Dependents'].replace(('0','1','2','3+'), ('0','1','2','3'), inplace=True)
```

```python
final['Credit_History'].replace((1.0, 0.0), ('1', '0'), inplace=True)
```

```python
final['Loan_Status'].value_counts()
```

```python
final['Loan_Status'].replace(('Y', 'N'), (1, 0), inplace = True)
```

```python
final['Gender'] = final['Gender'].astype('category')
final['Married'] = final['Married'].astype('category')
final['Dependents'] = final['Dependents'].astype('category')
final['Education'] = final['Education'].astype('category')
final['Self_Employed'] = final['Self_Employed'].astype('category')
final['Credit_History'] = final['Credit_History'].astype('category')
final['Property_Area'] = final['Property_Area'].astype('category')
```

```python
final.isnull().sum()
```

```python
final['Gender'].fillna(final['Gender'].mode()[0], inplace = True)
final['Married'].fillna(final['Married'].mode()[0], inplace = True)
final['Dependents'].fillna(final['Dependents'].mode()[0], inplace = True)
final['Self_Employed'].fillna(final['Self_Employed'].mode()[0], inplace = True)
final['Loan_Amount_Term'].fillna(final['Loan_Amount_Term'].mode()[0], inplace = True)
final['Credit_History'].fillna(final['Credit_History'].mode()[0], inplace = True)
```

```python
final.isnull().sum()
```

```python
table = final.pivot_table(values='LoanAmount', index='Self_Employed' ,columns='Education', aggfunc=np.median)

# Define function to return value of this pivot_table
def fage(x):
 return table.loc[x['Self_Employed'],x['Education']]

# Replace missing values
final['LoanAmount'].fillna(final[final['LoanAmount'].isnull()].apply(fage, axis=1), inplace = True)
```

```python
# Substituting median in last 2 percentile or ouliers 
median = final.loc[final['ApplicantIncome'] > 20000, 'ApplicantIncome'].median()
final.loc[final['ApplicantIncome'] > 20000, 'ApplicantIncome'] = np.nan
final['ApplicantIncome'].fillna(median, inplace=True)
```
